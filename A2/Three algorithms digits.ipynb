```
import time
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import svm, tree, datasets, metrics
from sklearn.neural_network import MLPClassifier 
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split, learning_curve, validation_curve
from sklearn.metrics import plot_roc_curve, classification_report, accuracy_score, f1_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.model_selection import GridSearchCV
import mlrose_hiive
```

# Load data

```
data = datasets.load_digits()
shapedData = data.data
X_train, X_test, y_train, y_test = train_test_split(shapedData, data.target, test_size = 0.5)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
one_hot = OneHotEncoder()
y_train = one_hot.fit_transform(y_train.reshape(-1, 1)).todense()
y_test = one_hot.transform(y_test.reshape(-1, 1)).todense()
iters = 1000
attempts = 100
act = 'relu'
rest = 0
schedule = mlrose_hiive.GeomDecay(init_temp=10, decay=0.0001, min_temp=0.001)
```

# **Random Hill Climb**

```
#@title Default title text
start_time = time.time()
rhc_nn = mlrose_hiive.NeuralNetwork(hidden_nodes = [40], activation=act, algorithm = 'random_hill_climb', max_iters = iters, early_stopping=True, max_attempts=attempts, learning_rate = 0.01, clip_max = 5, restarts=rest)
rhc_nn.fit(X_train, y_train)
rhc_nn_train_pred = rhc_nn.predict(X_train)
rhc_nn_train_accuracy = accuracy_score(y_train, rhc_nn_train_pred)
print (classification_report(y_train, rhc_nn_train_pred))
print ('RHC train accuracy: ', rhc_nn_train_accuracy)
print ("----------------------------------------------------------------------------------------------------------------")
rhc_nn_predicted = rhc_nn.predict(X_test)
rhc_nn_accuracy = accuracy_score(y_test, rhc_nn_predicted)
print (classification_report(y_test, rhc_nn_predicted))
print ('RHC test accuracy: ', rhc_nn_accuracy)
print("--- %s RHC seconds ---" % (time.time() - start_time))
```

# **Simulated Annealing**

```
start_time = time.time()
sa_nn = mlrose_hiive.NeuralNetwork(hidden_nodes = [40], activation=act, algorithm = 'simulated_annealing', schedule=schedule, max_iters = iters, early_stopping=True, max_attempts=attempts, learning_rate = 0.01, clip_max = 5)
sa_nn.fit(X_train, y_train)
sa_nn_train_pred = sa_nn.predict(X_train)
sa_nn_train_accuracy = accuracy_score(y_train, sa_nn_train_pred)
print (classification_report(y_train, sa_nn_train_pred))
print ('SA train accuracy: ', sa_nn_train_accuracy)
print ("----------------------------------------------------------------------------------------------------------------")
sa_nn_predicted = sa_nn.predict(X_test)
sa_nn_accuracy = accuracy_score(y_test, sa_nn_predicted)
print (classification_report(y_test, sa_nn_predicted))
print ('SA test accuracy: ', sa_nn_accuracy)
print("--- %s SA seconds ---" % (time.time() - start_time))
```

# **Genetic Alg**

```
start_time = time.time()
ga_nn = mlrose_hiive.NeuralNetwork(hidden_nodes = [40], activation=act, algorithm = 'genetic_alg', max_iters = iters, early_stopping=True, max_attempts=attempts, learning_rate = 0.01, pop_size=200, mutation_prob=0.1, clip_max = 5)
ga_nn.fit(X_train, y_train)
ga_nn_train_pred = ga_nn.predict(X_train)
ga_nn_train_accuracy = accuracy_score(y_train, ga_nn_train_pred)
print (classification_report(y_train, ga_nn_train_pred))
print ('GA train accuracy: ', ga_nn_train_accuracy)
print ("----------------------------------------------------------------------------------------------------------------")
ga_nn_predicted = ga_nn.predict(X_test)
ga_nn_accuracy = accuracy_score(y_test, ga_nn_predicted)
print (classification_report(y_test, ga_nn_predicted))
print ('GA test accuracy: ', ga_nn_accuracy)
print("--- %s GA seconds ---" % (time.time() - start_time))
```
